```python
import numpy as np
import pandas as pd
```

```python
r = 102303761
ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)
```

```python
df=pd.read_csv('/content/data.csv', encoding='latin1')
```

```python
df.head()
```

```python
x = df['no2'].dropna().values
```

```python
x
```

```python
z = x + ar * np.sin(br * x)
mu = np.mean(z)
sigma2 = np.mean((z - mu)**2)
```

```python
lam = 1 / (2 * sigma2)
c = 1 / np.sqrt(2 * np.pi * sigma2)

print("lambda =", lam)
print("mu =", mu)
print("c =", c)
```
